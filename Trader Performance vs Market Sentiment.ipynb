# **DATASET COLUMNS DESCRIPTION**

**Dataset 1: Market Sentiment**

Column	Description: -

- `timestamp` -	Unix timestamp representing the date of the sentiment value
- `value` -	Numerical Fear & Greed Index (0–100), where lower = fear, higher = greed
- `classification` -	Categorical label derived from index (e.g., Fear, Extreme Fear, Greed)
- `date` -	Human-readable date extracted from timestamp

**Dataset 2: Historical Trader Dataset**

- `Account` -	Unique identifier (wallet address) of the trader
- `Coin` -	Trading pair or asset symbol (e.g., BTC, ETH)
- `Execution Price` -	Price at which the trade was executed
- `Size Tokens` -	Quantity of tokens traded
- `Size USD` -	USD value of the trade
- `Side` -	Trade direction: BUY (long) or SELL (short)
- `Timestamp` -	Unix timestamp (in milliseconds) representing trade time (UTC)
- `Timestamp IST` -	Trade timestamp converted to Indian Standard Time (readable format)
- `Start Position` -	Position size before executing the trade
- `Direction` -	Trade intent (Buy/Sell direction classification)
- `Crossed` -	Boolean indicating whether the trade crossed the order book (taker trade)
- `Closed PnL` -	Profit or loss realized after closing the trade
- `Fee` -	Transaction fee paid for the trade
- `Transaction Hash` -	Unique blockchain transaction identifier
- `Order ID` -	Unique identifier for the order
- `Trade ID` -	Unique identifier for the executed trade

# **MERGED DATASET**

```python
# Merging the datasets
merged_df = trade.merge(senti, on='date', how='left')
```

```python
merged = merged_df[[
    'Account', 'coin', 'Execution Price', 'Size Tokens', 'Size USD', 'Side', 'Start Position', 'Direction', 'Closed PnL', 'Crossed', 'Fee', 'Order ID', 'date', 'value', 'classification'
]]
```

```python
merged_df['trade_type'] = merged_df.apply(
    lambda x: 'Open Long' if x['Side']=='BUY' and x['Direction']=='Buy'
    else 'Close Long' if x['Side']=='SELL' and x['Direction']=='Buy'
    else 'Open Short' if x['Side']=='SELL' and x['Direction']=='Sell'
    else 'Close Short',
    axis=1
)
```

```python
merged.isnull().sum()
```

```python
# Daily PnL per trader (or per account)
daily_pnl=merged.groupby(['Account', 'date'])['Closed PnL'].sum().reset_index()
daily_pnl
```

```python
# win rate
merged['is_win'] = merged['Closed PnL'] > 0
win_rate = merged.groupby('Account')['is_win'].mean() * 100
win_rate.sort_values(ascending=False)
```

```python
# average trade size per account
avg_trade_size = merged.groupby('Account')['Size USD'].mean()
avg_trade_size.sort_values(ascending=False)
```

```python
# average Trade size per day
avg_trade_size_per_day = merged.groupby('date')['Size USD'].mean()
avg_trade_size_per_day
```

```python
# number of trades per day
trades_per_day = merged.groupby('date').size().reset_index(name='num_trades')
trades_per_day
```

```python
# Long_short ratio
long_short=merged.groupby(['date', 'Side']).size().unstack().fillna(0)
long_short['long_short_ratio'] = long_short['BUY'] / (long_short['SELL'] + 1)
long_short.head()
```

**Performance vs Sentiment**

```python
# closed PnL vs Sentiment
sentiment_perf = merged_df.groupby('classification')['Closed PnL'].mean()
sentiment_perf
```

```python
# win_rate vs sentiment
win_rate_sentiment = merged.groupby('classification')['is_win'].mean()
win_rate_sentiment
```

- Traders exhibit lower average PnL and win rates during Fear periods compared to Greed periods.
- This suggests that market uncertainty and panic-driven decisions negatively impact performance.
- In contrast, Greed periods show relatively stable and higher profitability, indicating more favorable trading conditions.

**Behavior Change**

```python
behavior = merged_df.groupby('classification').agg({
    'Size USD': 'mean',
    'Closed PnL': 'mean',
    'Account': 'count'
})
behavior
```

During Greed phases, traders tend to:
- Increase trade size
- Take larger positions
- Trade more frequently

During Fear phases, traders:
- Reduce exposure
- Execute smaller trades
- Show more conservative behavior

**Frequent vs Infrequent traders**

```python
trade_counts = merged.groupby('Account').size()
threshold = trade_counts.median()

merged['trader_type'] = merged['Account'].map(
    lambda x: 'Frequent' if trade_counts[x] > threshold else 'Infrequent'
)

merged.groupby('Account')['trader_type'].first()
```

consistent winners vs inconsistent traders

```python
total_pnl = merged.groupby('Account')['Closed PnL'].sum()

merged['pnl_type'] = merged['Account'].map(
    lambda x: 'Winner' if total_pnl[x] > 0 else 'Loser'
)
merged.groupby('Account')['pnl_type'].first()
```

# **CONCLUSION**

This analysis demonstrates that market sentiment (Fear vs Greed) has a significant impact on both trader behavior and performance.

During Fear periods, traders tend to adopt a more cautious approach, characterized by reduced position sizes, lower trading activity, and more conservative decision-making. However, despite this caution, performance metrics such as PnL and win rate generally decline, likely due to increased market uncertainty and volatility.

In contrast, Greed periods are associated with increased trading activity, larger trade sizes, and a stronger directional bias toward long positions. These conditions often result in relatively better performance, suggesting that traders benefit from more stable and trend-driven markets.

The analysis also highlights that aggressive trading behavior (e.g., large position sizes or high exposure) leads to higher variability in outcomes, particularly during Fear phases where risk is amplified. Additionally, different trader segments (such as high-frequency vs low-frequency traders) respond differently to sentiment, indicating the importance of adaptive strategies.

Overall, the findings suggest that aligning trading strategies with market sentiment can improve risk-adjusted performance. In particular, adopting a defensive approach during Fear periods and a more opportunistic approach during Greed periods can help traders optimize outcomes while managing risk effectively.
